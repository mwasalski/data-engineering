```python
import pandas as pd
from azure.cosmosdb.table.tableservice import TableService

# Replace with your storage account name and access key
account_name = ""
access_key = ""
table_name = ""

# Create a table service client
table_service = TableService(account_name=account_name, account_key=access_key)

# Example: List tables
tables = table_service.list_tables()
for table in tables:
    print(table.name)
```

```python
# Query all entities from the table
entities = table_service.query_entities(table_name)
# Convert to DataFrame
df = pd.DataFrame([entity for entity in entities])
change_type = {'subregion':'string', 'region':'string', 'name':'string', 'intermediateregion':'string','alpha2':'string', 'alpha3':'string'}
rename = {'alpha3':'Country Code','intermediateregion':'Intermediate Region', 'name':'Country Name','region':'Region','subregion':'Sub Region'}

df_locationinfo = (
    df.drop(columns=['PartitionKey', 'RowKey', 'Timestamp'])
      .astype(change_type)
      .sort_values(by='alpha3', ascending=True)
      .rename(columns=rename)
      .drop(columns = ['midnightTimeStamp','objectType','etag'])
)
```
